In this section, I would be building a model to predict apartments prices in Buenos Aires.

+ Considering apartments in Buenos Aires in Capital Federal 

```python
import pandas as pd 
from glob import glob
```

```python
def wrangle(filepath):
    df = pd.read_csv(filepath)
    #Apartments in Capital Federal and cost less than 400_000
    df = df[df['property_type'] == 'apartment']
    df = df[df['place_with_parent_names'].str.contains('Capital Federal')]
    df = df[df['price_aprox_usd'] < 400_000]
    df[['lat', 'lon']] = df['lat-lon'].str.split(',', expand=True)
    df.drop(columns=['lat-lon'], inplace=True)
    
    low, high = df['surface_covered_in_m2'].quantile([0.05, 0.95])
    df = df[df['surface_covered_in_m2'].between(low, high)]
    return df
    
```

```python
files = glob('dataset/buenos-aires-real-estate-*.csv')
df_list = [wrangle(file) for file in files]
df = pd.concat(df_list)
df.head()
```
